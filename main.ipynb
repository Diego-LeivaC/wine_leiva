### **2) Support Vector Machine and Logistic Regression Implementation**

> Let's start with Support Vector Machine (SVM) from scratch

```python
# ------------------------------ SVM ------------------------------ #

# Labels: convert 0 to -1 and 1 to 1
def zeroOne(label):
    return np.where(label == 0, -1, 1)

y_train = zeroOne(training_data["quality"].values)

# Training SVM (vectorized)
def training_svm_vectorized(X, y, lr=0.0015, C=0.01, epochs=1000):
    n_samples, n_features = X.shape
    w = np.zeros(n_features)
    b = 0

    for _ in range(epochs):
        # Margin computation
        margins = y * (np.dot(X, w) - b)

        # Mask: errors (margin < 1)
        mask = margins < 1

        # Gradients
        dw = 2 * C * w - np.dot(X[mask].T, y[mask])
        db = -np.sum(y[mask])

        # Actualizations
        w -= lr * dw
        b -= lr * db

    return w, b

# Prediction
def test_svm(X, w, b):
    pred = np.dot(X, w) - b
    return np.where(pred > 0, 1, 0)

# Training
w_trained, b_trained = training_svm_vectorized(X_train, y_train)

# Prediction
y_pred = test_svm(X_test, w_trained, b_trained)

# Evaluation
accuracy = np.mean(y_pred == y_test_true)
print("SVM accuracy:", accuracy)

# 150 sample comparison between prediction vs real value in the test set
print("\nComparison of 150 samples(y_pred vs y_test_true):\n")
for i in range(150):
    print(f"Example {i+1}: Prediction = {y_pred[i]}, Real = {y_test_true[i]}")
```

> We continue with Logistic Regression (LR) from scratch

```python
# ------------------ Logistic Regression ------------------ #

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def training_lr(X, y, lr=0.00015, epochs=1000):
    n_samples, n_features = X.shape
    w = np.zeros(n_features)
    b = 0

    for _ in range(epochs):
        linear_model = np.dot(X, w) + b
        predictions = sigmoid(linear_model)

        # Gradient
        error = predictions - y
        dw = np.dot(X.T, error) / n_samples
        db = np.sum(error) / n_samples

        # Actualization
        w -= lr * dw
        b -= lr * db

    return w, b

def test_lr(X, w, b):
    probabilities = sigmoid(np.dot(X, w) + b)
    return (probabilities >= 0.5).astype(int)

y_train = training_data["quality"].values

# Training
w_lr, b_lr = training_lr(X_train, y_train)

# Prediction
y_pred_lr = test_lr(X_test, w_lr, b_lr)

# Evaluation
accuracy_lr = np.mean(y_pred_lr == y_test_true)
print("Logistic Regression accuracy:", accuracy_lr)
```

> The accuracy of the logistic regression is 0.6371 that is lower than the SVM one, then, I proceed with the hyperparameter tuning and accuracy estimation via 5-fold cross validation for Logistic Regression

```python
def k_fold_cross_validation_lr(X, y, learning_rates, epoch_numbers, k=5):
    np.random.seed(27)
    indices = np.arange(len(X))
    np.random.shuffle(indices)

    fold_size = len(X) // k
    best_score = 0
    best_parameters = None

    for lr in learning_rates:
        for epochs in epoch_numbers:
            fold_accuracies = []

            for i in range(k):
                start = i * fold_size
                end = start + fold_size if i < k - 1 else len(X)

                test_idx = indices[start:end]
                train_idx = np.concatenate((indices[:start], indices[end:]))

                X_train_cv = X[train_idx]
                y_train_cv = y[train_idx]
                X_test_cv = X[test_idx]
                y_test_cv = y[test_idx]

                w_cv, b_cv = training_lr(X_train_cv, y_train_cv, lr=lr, epochs=epochs)
                y_pred_cv = test_lr(X_test_cv, w_cv, b_cv)

                accuracy = np.mean(y_pred_cv == y_test_cv)
                fold_accuracies.append(accuracy)

            avg_accuracy = np.mean(fold_accuracies)
            print(f"LR = {lr}, Epochs = {epochs}, CV Accuracy = {avg_accuracy:.4f}")

            if avg_accuracy > best_score:
                best_score = avg_accuracy
                best_parameters = (lr, epochs)

    print(f"\nBest Hyperparameters for Logistic Regression: LR = {best_parameters[0]}, Epochs = {best_parameters[1]} with Accuracy = {best_score:.4f}")
    return best_parameters
```

```python
# Prepare the full data (without split in train and test)
X_full = data.drop("quality", axis=1).values
y_full = data["quality"].values

# Define the values that I want to prove
learning_rates = [0.0001, 0.00015, 0.001, 0.01]
epoch_numbers = [100, 500, 1000, 1500]

# Execute the cross validation with hyperparameter lookup
best_lr, best_epochs = k_fold_cross_validation_lr(X=X_full, y=y_full,
    learning_rates=learning_rates, epoch_numbers=epoch_numbers, k=5)
```

```python
# --------------------------- Final Training & Evaluation --------------------------- #

# Use the best hiperparameters found by cross-validation
best_lr = 0.0001
best_epochs = 1000

# Final training over the training set
w_final, b_final = training_lr(X_train, y_train, lr=best_lr, epochs=best_epochs)

# Prediction over the test set
y_pred_final = test_lr(X_test, w_final, b_final)

# Evaluation
final_accuracy = np.mean(y_pred_final == y_test_true)
print(f"Final Logistic Regression accuracy (lr={best_lr}, epochs={best_epochs}): {final_accuracy:.4f}")
```

### **3) Kernel Methods**

> Before implementing the Gaussian Kernel to the above mentioned algorithms, I normalize the X_train and X_test

```python
# Prepare data
X_train_raw = training_data.drop("quality", axis=1).values
X_test_raw = test_data.drop("quality", axis=1).values

# Compute mean and standard deviation on training data
mean = np.mean(X_train_raw, axis=0)
std = np.std(X_train_raw, axis=0)

# Normalize training data
X_train_norm = (X_train_raw - mean) / std

# Normalize test data using training mean and std
X_test_norm = (X_test_raw - mean) / std
```

> I proceed with the SVM

```python
# ------------------------------ SVM Kernel ------------------------------ #

def zeroOne(label):
    return np.where(label == 0, -1, 1)

def rbf_kernel(X1, X2, gamma=0.01):
    X1_sq = np.sum(X1**2, axis=1).reshape(-1, 1)
    X2_sq = np.sum(X2**2, axis=1).reshape(1, -1)
    squared_dist = X1_sq + X2_sq - 2 * np.dot(X1, X2.T)
    return np.exp(-1/2*gamma * squared_dist)

def training_svm_kernel_vectorized(X, y, lr=0.001, C=0.1, epochs=1500, gamma=0.01):
    n_samples = X.shape[0]
    alpha = np.zeros(n_samples)
    bias = 0
    K = rbf_kernel(X, X, gamma)

    for epoch in range(epochs):
        # Vectorize the margin computation for all j at once
        margins = y * (np.dot(K, alpha * y) + bias)  # Vector with margins for all j
        # Condition to update alphas: margin < 1
        violations = np.where(margins < 1)[0]  # Indices where margin < 1

        # Update alpha and bias only on violations
        alpha[violations] += lr
        #alpha[violations] += lr * (1 - margins[violations])
        bias += lr * np.sum(y[violations])

        # Regularization (shrink)
        alpha = np.maximum(0, alpha - lr * C * alpha)

    return alpha, bias

def predict_svm_kernel(X_test_norm, X_train_norm, y_train, alpha, bias, gamma=0.01):
    K_test = rbf_kernel(X_test_norm, X_train_norm, gamma)
    decision = np.dot(K_test, alpha * y_train) + bias
    return np.where(decision > 0, 1, 0)

# Prepare data
y_train = zeroOne(training_data["quality"].values)

# Train with vectorized inner loop
alpha_trained, bias_trained = training_svm_kernel_vectorized(X_train_norm, y_train)

# Predict and evaluate
y_pred_kernel = predict_svm_kernel(X_test_norm, X_train_norm, y_train, alpha_trained, bias_trained)
accuracy_kernel = np.mean(y_pred_kernel == y_test_true)
print("SVM Kernel (RBF) accuracy with vectorized inner loop:", accuracy_kernel)
```

> Now Kernel for the Logistic

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def rbf_kernel(X1, X2, gamma=0.1):
    X1_sq = np.sum(X1**2, axis=1).reshape(-1, 1)
    X2_sq = np.sum(X2**2, axis=1).reshape(1, -1)
    squared_dist = X1_sq + X2_sq - 2 * np.dot(X1, X2.T)
    return np.exp(-1/2*gamma  * squared_dist)

def training_lr_kernel(X, y, lr=0.0001, epochs=1000, gamma=0.1):
    n_samples = X.shape[0]
    w = np.zeros(n_samples)  # Weights in kernel space
    b = 0
    K = rbf_kernel(X, X, gamma)

    for _ in range(epochs):
        linear_model = np.dot(K, w) + b
        predictions = sigmoid(linear_model)
        error = predictions - y
        dw = np.dot(K.T, error) / n_samples
        db = np.sum(error) / n_samples

        w -= lr * dw
        b -= lr * db

    return w, b

def predict_lr_kernel(X_test_norm, X_train_norm, w, b, gamma=0.1):
    K_test = rbf_kernel(X_test_norm, X_train_norm, gamma)
    probs = sigmoid(np.dot(K_test, w) + b)
    return (probs >= 0.5).astype(int)

y_train = training_data["quality"].values

# Train Kernel logistic regression
w_kernel, b_kernel = training_lr_kernel(X_train_norm, y_train)
# Predict
y_pred_kernel_lr = predict_lr_kernel(X_test_norm, X_train_norm, w_kernel, b_kernel)

# Evaluate
accuracy_kernel_lr = np.mean(y_pred_kernel_lr == y_test_true)
print("Logistic Regression kernel accuracy:", accuracy_kernel_lr)
```

#### **4.1) Confusion matrix and metrics**

```python
def confusion_matrix_manual(y_true, y_pred):
    TP = np.sum((y_true == 1) & (y_pred == 1))
    TN = np.sum((y_true == 0) & (y_pred == 0))
    FP = np.sum((y_true == 0) & (y_pred == 1))
    FN = np.sum((y_true == 1) & (y_pred == 0))
    
    return np.array([[TN, FP],
                     [FN, TP]])

def accuracy_manual(y_true, y_pred):
    return np.mean(y_true == y_pred)

def precision_manual(y_true, y_pred):
    TP = np.sum((y_true == 1) & (y_pred == 1))
    FP = np.sum((y_true == 0) & (y_pred == 1))
    if TP + FP == 0:
        return 0
    return TP / (TP + FP)

def recall_manual(y_true, y_pred):
    TP = np.sum((y_true == 1) & (y_pred == 1))
    FN = np.sum((y_true == 1) & (y_pred == 0))
    if TP + FN == 0:
        return 0
    return TP / (TP + FN)

def f1_score_manual(y_true, y_pred):
    prec = precision_manual(y_true, y_pred)
    rec = recall_manual(y_true, y_pred)
    if (prec + rec) == 0:
        return 0
    return 2 * (prec * rec) / (prec + rec)

def evaluate_model_manual(y_true, y_pred, model_name="Model"):
    acc = accuracy_manual(y_true, y_pred)
    prec = precision_manual(y_true, y_pred)
    rec = recall_manual(y_true, y_pred)
    f1 = f1_score_manual(y_true, y_pred)
    cm = confusion_matrix_manual(y_true, y_pred)
    
    print(f"{model_name} Performance:")
    print("Confusion Matrix:")
    print("               Predicted 0   Predicted 1")
    print(f"Actual 0       {cm[0,0]:>12}   {cm[0,1]:>12}")
    print(f"Actual 1       {cm[1,0]:>12}   {cm[1,1]:>12}")
    print("\n")
    print(f"Accuracy:  {acc:.4f}")
    print(f"Precision: {prec:.4f}")
    print(f"Recall:    {rec:.4f}")
    print(f"F1-score:  {f1:.4f}")
    return acc, prec, rec, f1, cm
```

#### **4.2) Loss and Accuracy plot**

```python
# Kernel Logistic Regression with loss and accuracy history

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def training_lr_kernel_with_history(X, y, lr=0.0001, epochs=1500, gamma=0.1):
    n_samples = X.shape[0]
    w = np.zeros(n_samples)  # Weights in kernel space
    b = 0
    K = rbf_kernel(X, X, gamma)

    losses = []
    accuracies = []

    for _ in range(epochs):
        linear_model = np.dot(K, w) + b
        predictions = sigmoid(linear_model)

        # Binary cross-entropy loss
        loss = -np.mean(y * np.log(predictions + 1e-15) + (1 - y) * np.log(1 - predictions + 1e-15)) # Numerical safety trick to avoid taking the log of 0

        losses.append(loss)

        preds_binary = (predictions >= 0.5).astype(int)
        acc = np.mean(preds_binary == y)
        accuracies.append(acc)

        error = predictions - y
        dw = np.dot(K.T, error) / n_samples
        db = np.sum(error) / n_samples

        w -= lr * dw
        b -= lr * db

    return w, b, losses, accuracies

# Prepare y
y_train = training_data["quality"].values

# Train with history
w_kernel, b_kernel, losses_kernel, accs_kernel = training_lr_kernel_with_history(
    X_train_norm, y_train)
# Plot Loss and Accuracy curves
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(losses_kernel)
plt.title("Kernel Logistic Regression - Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.subplot(1, 2, 2)
plt.plot(accs_kernel)
plt.title("Kernel Logistic Regression - Training Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")

plt.show()
```

```python
# ----------------- Loss and Accuracy curve for Kernel SVM ----------------- #

def training_svm_kernel_vectorized_with_history(X, y, lr=0.001, C=0.1, epochs=1500, gamma=0.01):
    n_samples = X.shape[0]
    alpha = np.zeros(n_samples)
    bias = 0
    K = rbf_kernel(X, X, gamma)

    accs = []
    losses = []

    for epoch in range(epochs):
        margins = y * (np.dot(K, alpha * y) + bias)
        violations = np.where(margins < 1)[0]
        
        # Update alphas on violations
        alpha[violations] += lr

        # Update bias
        bias += lr * np.sum(y[violations])

        # Regularization (shrink)
        alpha = np.maximum(0, alpha - lr * C * alpha)

        # Compute hinge loss
        hinge_loss = np.mean(np.maximum(0, 1 - margins))
        losses.append(hinge_loss)

        # Accuracy on training set
        decision_values = np.dot(K, alpha * y) + bias
        preds_binary = np.where(decision_values > 0, 1, -1)
        acc = np.mean(preds_binary == y)
        accs.append(acc)

    return alpha, bias, accs, losses

# Prepare labels
y_train = zeroOne(training_data["quality"].values)

# Train with history
alpha_k, bias_k, accs_k, losses_k = training_svm_kernel_vectorized_with_history(
    X_train, y_train)

# Plot Loss and Accuracy curves
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(losses_k)
plt.title("SVM Kernel (RBF) Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.subplot(1,2,2)
plt.plot(accs_k)
plt.title("SVM Kernel (RBF) Training Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")

plt.show()
```
