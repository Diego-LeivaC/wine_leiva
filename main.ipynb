### **2) Support Vector Machine and Logistic Regression Implementation**

> Let's start with Support Vector Machine (SVM) from scratch

```python
# -------------------------- SVM -------------------------- #

# Labels: convert 0 to -1 and 1 to 1

def zeroOne(label):
    return np.where(label == 0, -1, 1)

y_train = zeroOne(training_data["quality"].values)
y_test_true = test_data["quality"].values  # to evaluate afterward

# Training SVM (vectorized)

def training_svm_vectorized(X, y, lr=0.0015, C=0.01, epochs=1000):
    n_samples, n_features = X.shape
    w = np.zeros(n_features)
    b = 0

    for _ in range(epochs):
        # Margin computation
        margins = y * (np.dot(X, w) - b)

        # Mask: errors (margin < 1)
        mask = margins < 1

        # Gradients
        dw = 2 * C * w - np.dot(X[mask].T, y[mask])
        db = -np.sum(y[mask])

        # Actualizations
        w -= lr * dw
        b -= lr * db

    return w, b

# Prediction

def test_svm(X, w, b):
    pred = np.dot(X, w) - b
    return np.where(pred > 0, 1, 0)

# Training
w_trained, b_trained = training_svm_vectorized(X_train, y_train)

# Prediction
y_pred = test_svm(X_test, w_trained, b_trained)

# Evaluation
accuracy = np.mean(y_pred == y_test_true)
print("SVM accuracy:", accuracy)
```

> We continue with Logistic Regression (LR) from scratch

```python
# LR
y_train = training_data["quality"].values

# -------------------------- Logistic Regression -------------------------- #

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def training_lr(X, y, lr=0.0001, epochs=1000):
    n_samples, n_features = X.shape
    w = np.zeros(n_features)
    b = 0

    for _ in range(epochs):
        linear_model = np.dot(X, w) + b
        predictions = sigmoid(linear_model)

        # Gradient
        error = predictions - y
        dw = np.dot(X.T, error) / n_samples
        db = np.sum(error) / n_samples

        # Actualization
        w -= lr * dw
        b -= lr * db

    return w, b

def test_lr(X, w, b):
    probabilities = sigmoid(np.dot(X, w) + b)
    return (probabilities >= 0.5).astype(int)

# Training
w_lr, b_lr = training_lr(X_train, y_train)

# Prediction
y_pred_lr = test_lr(X_test, w_lr, b_lr)

# Evaluation
accuracy_lr = np.mean(y_pred_lr == y_test_true)
print("Logistic Regression accuracy:", accuracy_lr)
```

> Then with the demostration of the hyperparameter tuning and accuracy estimation via 5-fold cross validation for Logistic Regression

```python
def k_fold_cross_validation_lr(X, y, learning_rates, epoch_numbers, k=5):
    np.random.seed(27)
    indices = np.arange(len(X))
    np.random.shuffle(indices)

    fold_size = len(X) // k
    best_score = 0
    best_params = None

    for lr in learning_rates:
        for epochs in epoch_numbers:
            fold_accuracies = []

            for i in range(k):
                start = i * fold_size
                end = start + fold_size if i < k - 1 else len(X)

                test_idx = indices[start:end]
                train_idx = np.concatenate((indices[:start], indices[end:]))

                X_train_cv = X[train_idx]
                y_train_cv = y[train_idx]
                X_test_cv = X[test_idx]
                y_test_cv = y[test_idx]

                w_cv, b_cv = training_lr(X_train_cv, y_train_cv, lr=lr, epochs=epochs)
                y_pred_cv = test_lr(X_test_cv, w_cv, b_cv)

                accuracy = np.mean(y_pred_cv == y_test_cv)
                fold_accuracies.append(accuracy)

            avg_accuracy = np.mean(fold_accuracies)
            print(f"LR = {lr}, Epochs = {epochs}, CV Accuracy = {avg_accuracy:.4f}")

            if avg_accuracy > best_score:
                best_score = avg_accuracy
                best_params = (lr, epochs)

    print(f"\nBest Hyperparameters for Logistic Regression: LR = {best_params[0]}, Epochs = {best_params[1]} with Accuracy = {best_score:.4f}")
    return best_params
```

```python
# Prepare the full data (without split in train and test)
X_full = data.drop("quality", axis=1).values
y_full = data["quality"].values

# Defining the values that I want to prove
learning_rates = [0.0001, 0.00015, 0.001, 0.01]
epoch_numbers = [100, 500, 1000, 2000]

# Executing the cross validation with hyperparameter lookup
best_lr, best_epochs = k_fold_cross_validation_lr(
    X=X_full,
    y=y_full,
    learning_rates=learning_rates,
    epoch_numbers=epoch_numbers,
    k=5
)
```

```python
# --------------------------- Final Training & Evaluation --------------------------- #

# Using the best hiperparameters found by cross-validation
best_lr = 0.0001
best_epochs = 1000

# Final training over the training set
w_final, b_final = training_lr(X_train, y_train, lr=best_lr, epochs=best_epochs)

# Prediction over the test set
y_pred_final = test_lr(X_test, w_final, b_final)

# Evaluation
final_accuracy = np.mean(y_pred_final == y_test_true)
print(f"Final Logistic Regression accuracy (lr={best_lr}, epochs={best_epochs}): {final_accuracy:.4f}")
```

```python
# Optional exploration
w_explore, b_explore = training_lr(X_train, y_train, lr=best_lr, epochs=2000)
y_pred_explore = test_lr(X_test, w_explore, b_explore)
accuracy_explore = np.mean(y_pred_explore == y_test_true)
print(f"🔎 Accuracy with epochs=2000: {accuracy_explore:.4f}")
```

### **3) Kernel Methods**

In this third part, I am going to implement the Gaussian Kernel to the above mentioned algorithms.
I proceed with the SVM

```python
def zeroOne(label):
    return np.where(label == 0, -1, 1)

def kernel(regressor1, regressor2, gamma=0.05):
    return np.exp(
        -gamma*(np.sum(regressor1**2, axis=1)[ : , np.newaxis]+np.sum(regressor2**2, axis=1)-2*np.dot(regressor1, regressor2.T)
    ))

def training_svm_kernel(regressor, label, learningRate = 0.001, regularizationParameter = 0.01, epochNumber = 500, gamma = 0.05):
    alpha = np.zeros(
        len(regressor)
    )
    bias=0
    kernelValue=kernel(regressor, regressor, gamma)
    for i in range(epochNumber):
        for j in range(len(regressor)):
            if (label[j]*(np.dot(alpha*label,kernelValue[ : ,j])+bias)<1):
                alpha[j] += learningRate
                bias += learningRate*label[j]
            alpha[j]=max(0, alpha[j]-learningRate*regularizationParameter*alpha[j])
    return alpha, bias

def test_svm_kernel(regressorTest, regressorTrain, labelTraining, alpha, bias, gamma=0.05):
    label = np.dot(
        kernel(regressorTest, regressorTrain, gamma),
        alpha*labelTraining
    ) + bias
    return [1 if i > 0 else 0 for i in label]

training_alpha, training_bias = training_svm_kernel(
    regressor=training_data.drop("quality", axis=1).values,
    label =zeroOne(training_data["quality"].values)
)
label_svm_kernel = test_svm_kernel(
    regressorTest= test_data.drop("quality",axis=1).values,
    regressorTrain=training_data.drop("quality", axis=1).values,
    labelTraining=zeroOne(training_data["quality"].values),
    alpha= training_alpha,
    bias=training_bias
)
```
