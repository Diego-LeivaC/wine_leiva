#### We continue with the data exploring, filtering and preprocessing

```python
red_data = pd.read_csv("data/winequality-red.csv", sep=";")
white_data = pd.read_csv("data/winequality-white.csv", sep=";")
```

```python
red_data = red_data.drop(columns=["type"])
```

```python
red_data.head()
```

```python
red_data = red_data.assign(type=0)
white_data = white_data.assign(type=1)
```

```python
red_data.head()
```

```python
data = pd.concat([red_data, white_data], ignore_index=True)
```

```python
data.shape
```

```python
data = data.drop_duplicates()
```

```python
data.shape
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data["quality"] = (data["quality"] >= 6).astype(int)
```

```python
print("\nBinary distribution (0 = 'bad', 1 = 'good'):")
data["quality"].value_counts()
```
